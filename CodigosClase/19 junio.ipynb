```python
#codigos de la clase de hoy
import matplotlib.pyplot as plt
from numpy import * 
import math as mt
```

```python
#fibonacci
phi = (1 + mt.sqrt(5))/2
f1,f2 = 1,1

Fibonacci = []
Time = []

t = 0.0

n = 1000000
print("Para n =",n)
print("Para phi =",phi)

while f2 <= n:
    
    fn = f2/f1
    Fibonacci.append(fn)
    Time.append(t)
    t = t + 1
    
    print("la diferencia con phi es: ",phi-fn,', fn+1/fn=',fn)
    
    f1,f2 = f2,f1+f2
```

```python
#grafiquemos
plt.figure(figsize=(12,4))

plt.axhline(y = phi, xmin = 0.0, color = 'blue', linestyle='--', linewidth = 1) 
plt.text(-3.5, phi, r'$\phi$', size=17) #(coordenaada x, coordenada y , simbolo en latex, tamaño)

plt.plot(Time,Fibonacci, alpha=0.9) #alpha es para el grosor del trazo
plt.plot(Time,Fibonacci,'r.') #grafica de punto rojos

plt.grid() #cuadricula de la grafica

plt.axhline(y = 1.0, xmin = 0.0, color = 'black') #line horizontal color negra donde empieza la grafica
```

```python
#Graficas en el plano
from pylab import scatter,xlabel,ylabel,xlim,ylim,show
from numpy import loadtxt

data = loadtxt("estrellas.txt",float) #me lo carga al buffer directo como flotante

x = data[:,0] #toma todos los datos de la primer columna de data los pone en la lista X
y = data [:,1] #''  '' ' ' ' ' ' ' ' '  segunda columna en data y los pone en la lista Y

plt.figure(figsize=(9,6))

scatter(x,y, c=y,cmap = 'hot') #este es el plt.plot
#este tipo de graficas se llaman de dispersion
#plt.plot(x,y,'r.')

plt.title("Scatter Plot de datos esperimentales",fontsize=15)

xlabel("Temperatura",fontsize=13)
ylabel("Magnitud",fontsize=13)
xlim(0, 13000)
ylim(-5,20)

plt.axhline(y = 0.0, xmin = 0.0, color = 'black')
plt.axvline(x = 500, ymin = 0.0, color = 'black')
plt.grid()
```

```python
#empezaremos creando una funcion que sea el mapeo logistico
def logistic(r,x):
    return r*x*(1-x)
    
```

```python
#probando 
r = 0.5
#x = np.arange(0,1,0.01)#los saltos
x = np.linspace(0,1,1000)#el numero de puntos 
y = logistic(r,x)
```

```python
#nota como el logistic solo nos dio valores entre cero y uno tambien
#vamos a graficarlo
plt.plot(x,y,'r')
```

```python
r = 0.5
x = 0.4
N = 100

#para guardar los valores de N
ens = []
xs = [] #estas son las poblaciones
for n in range(N):
    x = logistic(r,x)
    #print(x)
    xs.append(x)
    ens.append(n)
```
